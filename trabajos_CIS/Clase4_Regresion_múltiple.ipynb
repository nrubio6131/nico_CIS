```python
##importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston#.load_boston()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#importamos los datos
dfx = pd.DataFrame(load_boston().data, columns=(load_boston()).feature_names)
dfy = pd.DataFrame(load_boston().target)
```

```python
#Organizamos los datos
X=dfx.iloc[:,:]
X=X.to_numpy()
Y= dfy.iloc[:,0]
Y = Y[:, np.newaxis]
print("Dimensiones en X: {}".format(str(X.shape)))
print("Dimension en Y: {}".format(str(Y.shape)))
```

```python
plt.rcParams['figure.figsize'] = [15, 4]# para aumentar la iamgen den la visualizaciÃ³n de jupyter
f, [ax1, ax2, ax3] = plt.subplots(1, 3, sharey=True) 
ax1.scatter(X[:,0],Y)
ax1.set_xlabel("habitaciones por vivienda");ax1.set_ylabel("precio viviendas")
ax2.scatter(X[:,1],Y)
ax2.set_xlabel("edad media viviendas")
ax3.scatter(X[:,2],Y)
ax3.set_xlabel("distancia ponderadas a centros de trabajo ")
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.3, random_state=4)
```

```python
regresion_multiple = LinearRegression()
regresion_multiple.fit(X_train,Y_train)
```

```python
Y_pred= regresion_multiple.predict(X_test)
print('resultado del entrenamiento = {}'.format(regresion_multiple.score(X_train,Y_train)))
print('resultado con los datos de prueba = {}'.format(regresion_multiple.score(X_test,Y_test)))
```

```python
print('coeficientes : a = {}, b = {}'.format(regresion_multiple.coef_[0,:],regresion_multiple.intercept_))
```

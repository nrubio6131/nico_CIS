```python
##importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression


from sklearn.metrics import r2_score
```

```python
#importamos los datos
data1 = pd.read_csv('../../data/Salary_Data.csv',delimiter=',')
X = data1.iloc[:,0]
```

```python
#Organizamos los datos
X = data1.iloc[:,0]
X= X[:, np.newaxis]
Y= data1.iloc[:,1]
Y = Y[:, np.newaxis]
print("X <A침os de experiencia> = dimensiones -> " + str(X.shape))
print("Y       <Salario>       = dimensiones -> " + str(Y.shape))
plt.scatter(X,Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.3, random_state=49)
```

```python
#graficamos
plt.scatter(X,Y,color='blue', marker='^',alpha=0.5)
plt.title("Gr치fica 1")
plt.xlabel("A침os de experiencia")
plt.ylabel("Salario")
plt.show()
```

```python
f, [ax1, ax2] = plt.subplots(2, 1, sharey=True) 
ax1.scatter(X_train,Y_train,color='blue', marker='^',alpha=0.5)
ax2.scatter(X_test, Y_test ,color='blue', marker='^',alpha=0.5)
plt.show()
```

```python
regresion = LinearRegression()
regresion.fit(X_train,Y_train)
```

```python
Y_pred= regresion.predict(X_test)
print(Y_pred)
```

```python
plt.scatter(X_test, Y_test, alpha=0.5)
plt.plot(X_test, Y_pred, color="red")
plt.title("Sueldo vs a침os de experiencia - Conjunto de entrenamiento")
plt.xlabel ("A침os de experiencia")
plt.ylabel ("Sueldo $")
plt.show()
```

```python
print(r2_score(Y_test,Y_pred))
```

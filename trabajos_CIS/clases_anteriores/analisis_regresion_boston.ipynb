```python
##importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.linear_model import LinearRegression
```

```python
data=datasets.load_boston()
dfx = pd.DataFrame(data.data, columns=(data).feature_names)
dfy = pd.DataFrame(data.target, columns=['MEDV'])
X=dfx.to_numpy()
Y=dfy.to_numpy()

i3=0
plt.rcParams['figure.figsize'] = [16,10]# apra aumentar la iamgen den la visualizacin de jupyter
fig, axs = plt.subplots(4, 4)
for i1 in range(0,4):
    for i2 in range(0,4):
        axs[i1,i2].scatter(X[:,i3],Y)
        axs[i1,i2].set_ylabel('{}'.format(dfx.columns[i3]))
        axs[i1,i2].grid(linestyle='--',linewidth=.5)
        i3+=1
        if i3 == 13:
            break
for i1 in range(0,4):
    axs[3,i1].set_xlabel('{}'.format(dfy.columns[0]))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.3, random_state=4)
```

```python
regresion_multiple = LinearRegression()
regresion_multiple.fit(X_train,Y_train)
```

```python
Y_pred= regresion_multiple.predict(X_test)
print('resultado del entrenamiento = {}'.format(regresion_multiple.score(X_train,Y_train)))
print('resultado con los datos de prueba = {}'.format(regresion_multiple.score(X_test,Y_test)))
```

```python
print('coeficientes : a = {}, b = {}'.format(regresion_multiple.coef_[0,:],regresion_multiple.intercept_))
```

```python
#Organizamos los datos
X1=dfx.iloc[:,5]
X1 = X1[:, np.newaxis]
Y1= dfy.iloc[:,0]
Y1 = Y1[:, np.newaxis]
print("Dimensiones en X: {}".format(str(X1.shape)))
print("Dimension en Y: {}".format(str(Y1.shape)))
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1,test_size=.3, random_state=4)
regresion_simple = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=1)
regresion_simple.fit(X1_train,Y1_train)
Y1_pred= regresion_simple.predict(X1_test)
print('resultado del entrenamiento = {}'.format(regresion_simple.score(X1_train,Y1_train)))
print('resultado con los datos de prueba = {}'.format(regresion_simple.score(X1_test,Y1_test)))
print('coeficientes : a = {}, b = {}'.format(regresion_simple.coef_[0,:],regresion_simple.intercept_))
```

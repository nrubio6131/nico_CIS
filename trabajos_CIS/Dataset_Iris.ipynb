```python
##importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
print(load_iris().data.shape)
print(load_iris().target.shape)
print(load_iris().feature_names)
print(load_iris().target_names)
```

```python
dfx = pd.DataFrame(load_iris().data[:], columns=(load_iris().feature_names))
dfy = pd.DataFrame(load_iris().target[:],columns=['target'])
df=pd.concat([dfx, dfy], axis=1)
for i in range(3):
    df['target']=df['target'].replace(i,load_iris().target_names[i])
```

```python
datosx=[0,2]
X=load_iris().data
Y=load_iris().target
plt.scatter(X[:, datosx[0]], X[:, datosx[1]], c=Y)

plt.show()
```

```python
sns.pairplot(df, hue='target', size=3)
```
